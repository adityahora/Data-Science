1. Target Variable Analysis (Match Outcome)

Identify:
● Home wins, away wins, and draws distribution


● Most frequent outcome type


Compare: Home win vs away win frequency

```python
import pandas as pd 

df = pd.read_csv("D:\\project hour\\World Cup\\results.csv")

teams = ["Canada","Mexico","United States","Australia","Iran","Japan","Jordan","South Korea","Qatar","Saudi Arabia","Uzbekistan","Algeria","Cabo Verde","Côte d'Ivoire","Egypt","Ghana","Morocco","Senegal","South Africa","Tunisia","Argentina","Brazil","Colombia","Ecuador","Paraguay","Uruguay","New Zealand","Austria","Belgium","Croatia","England","France","Germany","Netherlands","Norway","Portugal","Scotland","Spain","Switzerland","Sweden"]

df = df[df['home_team'].isin(teams) | df['away_team'].isin(teams)]

print(df.head(10))

```

```python


home_goals = [i for i in df['home_score']]
away_goals = [i for i in df['away_score']]

print(home_goals)
print(away_goals)

home_win = (df['home_score'] > df['away_score']).sum()
away_win = (df['home_score'] < df['away_score']).sum()
draw = (df['home_score'] == df['away_score']).sum()


print("Home wins : ",home_win)    
print("Away wins : ",away_win)
print("Draws : ",draw)    

import numpy as np

arr = np.array([home_win, away_win, draw])
percentages = (arr / arr.sum() * 100)

dict = {}

dict['Home wins'] = f"{round(percentages[0],2)}%"
dict['Away wins'] = f"{round(percentages[1],2)}%"
dict['Draws'] = f"{round(percentages[2],2)}%"

for key, value in dict.items():
    print(key," : ", value)


comparison = round((percentages[0]-percentages[1]),2)
print(f"Home wins is {comparison}% more than Away wins.")
```

Here, the frequency of Home wins is 20.77% greater than Away wins.

2. Temporal Analysis (Match Date)


Analyze:

● Matches played per year

● Outcome trends over time

Identify:

● Years with most matches

● Any change in home/away win trends over time

```python

import matplotlib.pyplot as plt
a = pd.to_datetime(df['date'])
df['year'] = a.dt.year
# print(df)

year = 1872

matches_per_year = {'Year' : [],
                    'Match Count' : [],
                    'Home Wins': [],
                    'Away Wins' : [],
                    'Draws' : []}

for i in range(year, 2026):
    yearly_df = df[df['year'] == year]

    matches_per_year['Year'].append(year)
    matches_per_year['Match Count'].append(len(yearly_df))

    matches_per_year['Home Wins'].append((yearly_df['home_score'] > yearly_df['away_score']).sum())
    matches_per_year['Away Wins'].append((yearly_df['home_score'] < yearly_df['away_score']).sum())
    matches_per_year['Draws'].append((yearly_df['home_score'] == yearly_df['away_score']).sum())
    year += 1


print(df := pd.DataFrame(matches_per_year))
df['Home Advantage'] = (
    df['Home Wins'] - df['Away Wins']
) / df['Match Count']

df.to_csv("Matches_Count.csv")


plt.figure()
plt.figure(figsize=(16, 6))
plt.plot(df['Year'], df['Home Advantage'])
plt.axhline(0)  # neutral line
plt.xlabel('Year')
plt.ylabel('Home Advantage')
plt.title('Home Advantage Over Time')
plt.show()
```

5. Tournament Analysis

Analyze:

● Number of matches per tournament

● Outcome distribution by tournament

Identify:

● Tournaments with more draws

● High-scoring tournaments

```python
matches_per_tournament = df['tournament'].value_counts()
print("Number of matches per tournament:")
print(matches_per_tournament)

def match_outcome(row):
    if row['home_score'] > row['away_score']:
        return 'Win'
    elif row['home_score'] < row['away_score']:
        return 'Loss'
    else:
        return 'Draw'

df['home_outcome'] = df.apply(match_outcome, axis=1)

outcome_distribution = df.groupby('tournament')['home_outcome'].value_counts()
print("\nOutcome distribution by tournament:")
print(outcome_distribution)

draws_per_tournament = df[df['home_outcome']=='Draw']['tournament'].value_counts()
print("\nTournaments with more draws:")
print(draws_per_tournament)

avg_goals_per_tournament = df.groupby('tournament')['total_goals'].mean()
high_scoring_tournaments = avg_goals_per_tournament[avg_goals_per_tournament > 3]
print("\nHigh-scoring tournaments:")
print(high_scoring_tournaments)
```
